# Ejercicio 1: Cálculo de ganancia de los repetidores analógicos

Asuma que cada repetidor analógico puede transmitir como máximo con una energía promedio $\xi$. Determine el valor de $A$ y las ganancias $G_{2},\dots,G_{n}$ para satisfacer ese requisito.

Exprese las ganancias en función de la relación señal a ruido a la salida del primer repetidor:

$$
SNR_{1} = h^{2}\,\frac{\xi}{\sigma^{2}}
$$

Definimos la relación señal a ruido (SNR) como el cociente entre la energía de la señal de interés y la energía (varianza) del ruido:

$$
SNR = \frac{\textrm{Energía de la señal de interés}}{\textrm{Energía del ruido}}
$$

$$
SNR_{1} = \frac{\mathbb{E}[(h\,X_1)^2]}{\mathbb{E}[(W_1)^2]} = \frac{h^2\,\mathbb{E}[(X_1)^2]}{\sigma^2}  = \frac{h^2\,\xi}{\sigma^2}
$$

Conociendo la energia transmitida en cada etapa es posible determinar la expesion de la ganacia en funcion de la SNR de cada etapa. 
Primero, hallamos la energia del simbolo trasmitida $\mathbb{E[\mathbf{X_i^2}]}$. La señal $\mathbf{X_i}$ es una variable aleatoria de tipo Bernoulli cuya distribucion sera:
$$
X_{1} = \left\{\begin{array}{ll}
        A & \textrm{con } p = 1/2\\
        -A & \textrm{con } p=1/2
      \end{array}\right.
$$

Por lo tanto, si desarrollamos la expresion de la energia ddel simbolo segun esta variable:
$$
\begin{aligned}
\mathbb{E}\left[X_{1}^2\right] &= A^{2} \, \mathbb{P}\left[X_{1} = A\right] + (-A)^{2} \, \mathbb{P}\left[X_{1} = -A\right]\\
&= A^{2} \, \frac{1}{2} + (-A)^{2} \, \frac{1}{2} \\
&= A^{2}
\end{aligned}
$$

Considerando la condición de que tanto el bloque $Tx$ como los $G_{i}$ transmitan con una energía promedio $\xi$ (con $\xi$ constante):

$$
\mathbb{E}\left[X_{i}^{2}\right] \leq \xi \qquad \forall i \in[1,n]
$$

De hecho, para todos los cálculos vamos a considerar a condición más favorable: $\mathbb{E}\left[X_{i}^{2}\right] = \xi$ para todo $i$. Por lo tanto, es posible determinar el valor de A:
$$ \mathbf{A}=\mathbf{\sqrt{\xi}}$$

Por otro lado, podemos plantear las ecuaciones del repetidor analogico para hallar una relacion directa entre la variable aleatoria $\mathbf{X_i}$ y el factor de ganacia en cada etapa:
$$
\left.\begin{array}{l}
        Y_{i} = h\,X_{i} + W_{i}\\
        X_{i+1} = G_{i+1}\,Y_{i}
      \end{array}\right\} \Rightarrow X_{i+1} = G_{i+1}\,\big(h\,X_{i} + W_{i}\big) 
$$

Si aplico el operador  cuadratico y luego el operador esperanza a la relacion hallada en el punto anterior obtengo:
$$\mathbb{E}[X_{i+1}^2]=\mathbb{E}[\left(G_{i+1}\,\big(h\,X_{i} + W_{i}\big) \right)^2]$$

$$\mathbb{E}[X_{i+1}^2]=G_{i+1}^2\,\big(h^2\mathbb{E}[\,X_{i}^2] +2h\,\mathbb{E}[\,X_{i}\,W_{i}]+ \mathbb{E}[W_{i}^2]\big)$$



Pero, si asumo que los ruidos son independientes a las señales, y que $W_i \sim \mathcal{N}(0, \sigma^2)$ con media nula, entonces puedo decir que:
$$\mathbb{E}[2\,X_{i}\,W_{i}]=2\left( Cov(X_i\,W_i)-\mathbb{E}[X_i]\,\mathbb{E}[W_i]\right)=0$$

Entonces, sigo con el desarrollo:

$$1=G_{i+1}^2\,\frac{\big(h^2\mathbb{E}[\,X_{i}^2] + \mathbb{E}[W_{i}^2]\big)}{\mathbb{E}[X_{i+1}^2]}$$

$$1=G_{i+1}^2\,\left(\frac{h^2\mathbb{E}[\,X_{i}^2]}{\mathbb{E}[X_{i+1}^2]} +\frac{\mathbb{E}[W_{i}^2]}{\mathbb{E}[X_{i+1}^2]}\right)$$

$$1=G_{i+1}^2\,\left(h^2 +\frac{h^2}{SNR_i}\right)$$

$$1=G_{i+1}^2\,\left(\frac{h^2\,SNR_i+h^2}{SNR_i}\right)$$

$$G_{i+1}^2=\left(\frac{SNR_i}{h^2\,SNR_i+h^2}\right)$$

$$G_{i+1}=\sqrt{\left(\frac{SNR_i}{h^2\,SNR_i+h^2}\right)} \\
G_{i+1}=\dfrac{1}{h}\sqrt{\frac{SNR_i}{SNR_i+1}}
$$





# Ejercicio 2: Probabilidad de error del sistema analógico

1. Exprese $Y_{n}$ en función de $X_{1}$, las ganancias y los ruidos $W_{1},\dots,W_{n}$. ¿Qué distribución tiene los términos asociados sólo al ruido?
2. Calcule la relación señal a ruido en la última etapa $\rho_{n}$.
3. Calcule la probabilidad de error promedio en función de la SNR de la última etapa $\rho_{n}$.

## 2.1) Salidas $Y_n$ en función de $X_1, W_1, ..., W_n$

Definiendo $Y_i = h X_i + W_i$ con $X_i = G_i Y_{i-1}$, se obtiene para $Y_3$:
$$Y_3 = h(G_3(hG_2(hX_1 + W_1) + W_2) + W_3) $$
lo cual se puede reescribir de forma generica como:
$$ Y_i = X_i h^i \prod_{k=2}^i G_k + \sum_{j=1}^i \left[ h^{i-j}W_{j+1} \prod_{k=j+1}^i G_k \right] $$

Los terminos de la ecuación anterior asociados al ruido son:
$$ \phi_i =  \sum_{j=1}^i \left[ h^{i-j}W_{j+1} \prod_{k=j+1}^i G_k \right] $$ 

Dichos terminos se comportan como una normal, de forma tal que:
$$ \phi_i \sim \mathcal{N}\left( 0, \sum_{j=1}^i \left[ \sigma^2h^{2(i-j)} \prod_{k=j+1}^i G_k^2 \right] \right) $$

## 2.2) Relación señal a riudo en la última etapa $\rho_n$
La señal en la última etapa $\rho_n$ se puede expresar como
$$ Y_n = \underbrace{X_1 h^n \prod_{k=2}^nG_k}_{\chi_n} + \underbrace{\sum_{j=1}^n \left[ h^{n-j} W_{j+1} \prod_{k=j+1}^n G_k \right]}_{\phi_n}$$

La $SNR_n$ se obtiene mediante la división de la autocorrelación de cada parte de la ecuación, aislando la parte útil de la señal y la parte correspondiente al ruido de la misma. 

$$ SNR_n = \dfrac{\mathbb{E}[\chi_n^2]}{\mathbb{E}[\phi_n^2]} $$ 
donde:
$$
\mathbb{E}[\chi_n^2] = \xi h^{2n} \prod_{k=2}^n G_k^2 \\
\mathbb{E}[\phi_n^2] = \sum_{j=1}^n \left[ \sigma^2 h^{2(n-j)} \prod_{k=j+1}^n G_k^2 \right] \\
\Rightarrow SNR_n = \dfrac{\xi h^{2n} \prod_{k=2}^n G_k^2}{\sum_{j=1}^n \left[ \sigma^2 h^{2(n-j)} \prod_{k=j+1}^n G_k^2 \right]} = 
\dfrac{\xi h^2}{\sigma^2} \dfrac{ h^{2n-2} \prod_{k=2}^n G_k^2}{\sum_{j=1}^n \left[  h^{2(n-j)} \prod_{k=j+1}^n G_k^2 \right]}
$$
donde $$ SNR_1 = \dfrac{\xi h^2}{\sigma^2} $$

Continuando con el desarrollo:
$$
SNR_n = SNR_1 \dfrac {h^{2n-2}G_2^2G_3^2 \cdot \cdot \cdot G_n^2}{h^{2(n-1)}G_2^2G_3^2 \cdot \cdot \cdot G_n^2 + h^{2(n-2)}G_3^2G_4^2 \cdot \cdot \cdot G_n^2 + \cdot \cdot \cdot } \\
= SNR_1 \dfrac {h^{2n-2}}{h^{2(n-1)} \dfrac{G_2^2G_3^2 \cdot \cdot \cdot G_n^2}{G_2^2G_3^2 \cdot \cdot \cdot G_n^2} + h^{2(n-2)}\dfrac{G_3^2G_4^2 \cdot \cdot \cdot G_n^2}{G_2^2G_3^2 \cdot \cdot \cdot G_n^2} + \cdot \cdot \cdot } \\
= SNR_1 \dfrac{h^{2(n-1)}}{h^{2(n-1)} + h^{2(n-2)}G_2^{-2} + h^{2(n-3)}G_2^{-2}G_3^{-2} + \cdot \cdot \cdot } \\
= SNR_1 \dfrac{1}{\dfrac{h^{2(n-1)}}{h^{2(n-1)}} + \dfrac{h^{2(n-2)}}{h^{2(n-1)}}G_2^{-2} + \dfrac{h^{2(n-3)}}{h^{2(n-1)}}G_2^{-2}G_3^{-2} + \cdot \cdot \cdot} \\
= SNR_1 \dfrac{1}{1 + h^{-2}G_2^{-2} + h^{-4}G_2^{-2}G_3^{-2} + \cdot \cdot \cdot } \\
\Rightarrow SNR_n = \dfrac{SNR_1}{\sum_{j=1}^n \left[ h^{-2(j-1)} \prod_{k=2}^n G_k^{-2} \right]}
$$

Del punto anterior, se sabe que $G_k = \dfrac{1}{h} \sqrt{\dfrac{SNR_1}{SNR_1+1}} $, por lo que:
$$
SNR_n = \dfrac{SNR_1}{\sum_{j=1}^n \left[ h^{-2(j-1)} \prod_{k=2}^n \left(\dfrac{1}{h} \sqrt{\dfrac{SNR_1}{SNR_1+1}} \right) ^{-2} \right]} \\
= \dfrac{SNR_1}{\sum_{j=1}^n \left[ h^{-2(j-1)} \prod_{k=2}^n h^2 \dfrac{SNR_1+1}{SNR_1} \right]} \\
= \dfrac{SNR_1}{\sum_{j=1}^n \left[ h^{-2(j-1)} \left( h^2 \dfrac{SNR_1+1}{SNR_1} \right) ^{j-1} \right] } \\
= \dfrac{SNR_1}{\sum_{j=1}^n \left[ h^{-2(j-1)} h^{2(j-1)} \left( \dfrac{SNR_1+1}{SNR_1} \right) ^{j-1} \right]} \\
= \dfrac{SNR_1}{\sum_{j=1}^n \left[\left( \dfrac{SNR_1+1}{SNR_1} \right) ^{j-1} \right]}
$$

Mediante un cambio de variable, se puede simplificar la ecuación anterior. Tomando $k = j-1$, resulta:
$$
SNR_n = \dfrac{SNR_1}{\sum_{k=0}^{n-1} \left[\left( \dfrac{SNR_1+1}{SNR_1} \right) ^k \right]}
$$

Utilizando la serie geometrica parcial, $\sum_{n=0}^N r^n = \dfrac{1-r^{N+1}}{1-r} $, se llega a:

$$
SNR_n = \dfrac{SNR_1}{\dfrac{1 - \left(\frac{SNR_1+1}{SNR_1}\right)^n}{1 - \frac{SNR_1+1}{SNR_1}}} 
\\
= \dfrac{SNR_1 \left(1 - \frac{SNR_1+1}{SNR_1}\right)}{1 - \left(\frac{SNR_1+1}{SNR_1} \right)^n}
$$

Desarrollando:

$$ 
      \left\{\begin{array}{ll}
        SNR_1 \left(1 - \frac{SNR_1+1}{SNR_1}\right) = -1\\
        1 - \left(\frac{SNR_1+1}{SNR_1} \right)^n = \dfrac{SNR_1^n - (SNR_1+1)^n}{SNR_1^n}
      \end{array}\right.
$$

Finalmente, 
$$
SNR_n = \dfrac{SNR_1^n}{(SNR_1+1)^n - SNR_1^n}
$$


## 2.3) Probabilidad de error promedio en función de la $SNR$ de la última etapa $\rho_n$

La probabilidad de error se puede calcular utilizando probabilidades totales y considerando que el receptor determina el valor de $\hat{X_1}$ como:
$$ 
\hat{X_1} = \left\{\begin{array}{ll}
        A & \textrm{si } Y_n >0\\
        -A & \textrm{si } Y_n \leq 0
      \end{array}\right.
$$

Se calcula la probabilidad usando la formula de probabilidades totales de manera que:
$$
\mathbb{P}_e = \mathbb{P}[\hat{X}_1 \neq X_1] \\
= \mathbb{P}[\hat{X}_1 \neq X_1 | X_1 = A] \mathbb{P}[X_1 = A] + \mathbb{P}[\hat{X}_1 \neq X_1 | X_1 = -A] \mathbb{P}[X_1 = -A] \\
= \mathbb{P}[\hat{X}_1 =-A | X_1 = A] \mathbb{P}[X_1 = A] + \mathbb{P}[\hat{X}_1 = A | X_1 = -A] \mathbb{P}[X_1 = -A] \\
= \mathbb{P}[Y_n < 0 | X_1 = A]\mathbb{P}[X_1 = A] + \mathbb{P}[Y_n \geq 0 | X_1 = -A]\mathbb{P}[X_1 = -A]
=\frac{1}{2}\left[ \mathbb{P}[X_1 = A] + \mathbb{P}[Y_n \geq 0 | X_1 = -A]\right]
$$

Anteriormente se encontró que:
$$
Y_n = \underbrace{X_1 h^n \prod_{k=2}^nG_k}_{\chi_n} + \underbrace{\sum_{j=1}^n \left[ h^{n-j} W_{j+1} \prod_{k=j+1}^n G_k \right]}_{\phi_n} \\
\phi_n \sim \mathcal{N}\left(0, \sum_{j=1}^n\left[ \sigma^2 h^{2(n-j)} \prod_{k=j+1}^n G_k^2 \right] \right) 
$$

si $G_k = \dfrac{1}{h}\sqrt{\dfrac{SNR}{SNR+1}} $,

$$
\Rightarrow \phi_n \sim \mathcal{N}\left(0, \sum_{j=1}^n\left[ \sigma^2 h^{2(n-j)} \prod_{k=j+1}^n \left( \dfrac{1}{h}\sqrt{\dfrac{SNR}{SNR+1}} \right)^2 \right] \right) \\
\sim \mathcal{N}\left(0, \underbrace{\sum_{j=1}^n\left[ \sigma^2 h^{2(n-j)} \prod_{k=j+1}^n \dfrac{1}{h^2}\dfrac{SNR}{SNR+1} \right]}_{(1)} \right)
$$
donde
$$
(1) = \sum_{j=1}^n\left[ \sigma^2 h^{2(n-j)} \prod_{k=j+1}^n \dfrac{1}{h^2}\dfrac{SNR_j}{SNR_j+1} \right] \\
= \sigma^2  \sum_{j=1}^n\left[ \dfrac{SNR_1}{SNR_1 + 1} \right]^{n-j} = \sigma^2 (1 + SNR_1) \left[ 1 - \left( \dfrac{SNR_1}{SNR_1+1} \right)^n \right]\\
\Rightarrow \phi_n \sim \mathcal{N}\left(0, \sigma^2 (1 + SNR_1) \left[ 1 - \left( \dfrac{SNR_1}{SNR_1+1} \right)^n \right]\right)
$$

Por lo tanto, al condicionar la variable a $X_1 = \pm A$ queda:
$$
Y_{n|x_1=\pm A} \sim \mathcal{N} \left( \pm A h^n \prod_{k=2}^nG_k, \sigma^2 (1 + SNR_1) \left[ 1 - \left( \dfrac{SNR_1}{SNR_1+1} \right)^n \right] \right)
$$
donde $h^n\prod_{k=2}^n G_k= h\left( \dfrac{SNR_1}{SNR_1+1} \right)^{\frac{n-1}{2}}$
$$
\Rightarrow Y_{n|x_1=\pm A} \sim \mathcal{N} \left( \pm A h\left( \dfrac{SNR_1}{SNR_1+1} \right)^{\frac{n-1}{2}}, \sigma^2 (1 + SNR_1) \left[ 1 - \left( \dfrac{SNR_1}{SNR_1+1} \right)^n \right] \right)
$$

Como se pide la probabilidad en función de la SNR de la última etapa $\rho_n$, y 
$$
SNR_1 = \dfrac{1}{\left( \frac{1}{SNR_n} + 1 \right)^{1/n} - 1}
$$
resulta:
$$
Y_{n|x_1=\pm A} \sim \mathcal{N} \left( \pm A h\left( \dfrac{\frac{1}{\left( \frac{1}{SNR_n} + 1 \right)^{1/n} - 1}}{\frac{1}{\left( \frac{1}{SNR_n} + 1 \right)^{1/n} - 1}+1} \right)^{\frac{n-1}{2}}, \sigma^2 \left(1 + \frac{1}{\left( \frac{1}{SNR_n} + 1 \right)^{1/n} - 1} \right) \left[ 1 - \left( \dfrac{\frac{1}{\left( \frac{1}{SNR_n} + 1 \right)^{1/n} - 1}}{\frac{1}{\left( \frac{1}{SNR_n} + 1 \right)^{1/n} - 1}+1} \right)^n \right] \right)
$$

Sabiendo que 
$$
\mathbb{P}[Y_n \geq 0 | X_1 = -A] = 1 - \mathbb{P}[Y_n < 0 | X_1 = -A]
$$

Al normalizar la función de distribución teniendo en cuenta que 
$$
\mathbb{P}[X_1 = A] = \mathbb{P}[X_1 = -A] = \frac{1}{2}
$$
resulta:
$$
\mathbb{P}_e = \frac{1}{2} \Phi \left( -\frac{\mu_{Y_n|X_1=A}}{\sigma_{Y_n|X_1=A}} \right) + \frac{1}{2} \left[ 1 - \Phi \left( -\frac{\mu_{Y_n|X_1=-A}}{\sigma_{Y_n|X_1=-A}} \right) \right]
$$

Considerando tambien que:
$$
\left\{\begin{array}{ll}
        \mu_{Y_n|X_1=A} = -\mu_{Y_n|X_1=-A} \\
        \sigma_{Y_n|X_1=A} = \sigma_{Y_n|X_1=-A}
      \end{array}\right.
$$
Se llega a:
$$
\mathbb{P}_e = \frac{1}{2} \left[ 1 - \Phi \left( \frac{\mu_{Y_n|X_1=A}}{\sigma_{Y_n|X_1=A}} \right) \right] + \frac{1}{2} \left[ 1 - \Phi \left( \frac{\mu_{Y_n|X_1=A}}{\sigma_{Y_n|X_1=A}} \right) \right] \\
\Rightarrow \mathbb{P} = 1 - \Phi \left( \frac{\mu_{Y_n|X_1=-A}}{\sigma_{Y_n|X_1=-A}} \right)
$$

Sabiendo que:
$$
\left\{\begin{array}{ll}
       \mu_{Y_n|X_1=-A} = Ah^n \prod_{k=2}^nG_k \\
        SNR_n = \dfrac{\mathbb{E}[\chi_n^2]}{\mathbb{E}[\phi_n^2]} = \dfrac{\xi h^{2n}\prod_{k=2}^nG_k^2}{\sigma_{Y_n|X_1=A}^2} \Rightarrow \sigma_{Y_n|X_1=A} = \dfrac{\sqrt{\xi}h^n\prod_{k=2}^nG_k}{\sqrt{SNR_n}}
      \end{array}\right. \\
$$

Finalmente, la probabilidad de error resulta:
$$
\mathbb{P}_e = 1 - \Phi \left( \dfrac{\mu_{Y_n|X_1=-A}}{\sigma_{Y_n|X_1=-A}} \right) = 1 - \Phi \left( \dfrac{Ah^n \prod_{k=2}^nG_k}{\dfrac{\sqrt{\xi}h^n\prod_{k=2}^nG_k}{\sqrt{SNR_n}}} \right) \underbrace{=}_{A=\sqrt{\xi}} 1 - \Phi{\sqrt{SNR_n}}
$$

Definiendo 
$$
Q(X) = \mathbb{P}[X \geq x] = 1 - \Phi(x)
$$
considerando $X \sim \mathcal{N}(0,1)$, se llega a que la probabilidad de error para un sistema analogico resulta:
$$
\mathbb{P}_e = Q(\sqrt{SNR_n})
$$

# Ejercicio 3

Grafico de las probabilidades de error del sistema digital y analogico en funcion de $SNR$ para el rango de -5 dB a 30 dB con pasos de 1 dB.
Dichas curvas se grafican para n entre 1 y 25 con pasos de 4

```
import numpy as np
import scipy as sp
import scipy.stats as sps
import matplotlib.pyplot as plt
import math as m
```

## Funcion de probabilidad de error

```
def prob_error_digital(snr,n):
  return 0.5*(1-(1-2*sps.norm.sf(np.sqrt(snr)))**n)

def prob_error_analogico(snr,n):
  return sps.norm.sf(np.sqrt(1 / ( (1+(1/snr_times))**n - 1)))
```

## Graficos


```
snr_db = np.arange(-5,30,1)
snr_times = 10**(snr_db/10)
enes = np.arange(1,25+4,4)
```

```
plt.figure(figsize=(25,10))

for n in enes:
  plt.semilogy(snr_db,prob_error_digital(snr_times,n),'-o',label = 'prob. de error sistema digital para $n={}$'.format(n))
  plt.semilogy(snr_db,prob_error_analogico(snr_times,n),'-',label = 'prob. de error sistema analogico para $n={}$'.format(n))

plt.xticks(snr_db)
plt.legend(loc='best')
plt.xlabel("SNR [dB]")
plt.ylabel("Probabilidad de error")
plt.xlim([-5,30])
plt.ylim([10e-7,1])
plt.grid(True)
plt.show()
```

## 3.2) Análisis de Resultados
En el gráfico, se puede apreciar que las curvas dependen tanto de la SNR como de la cantidad de etapas de los sistemas.
En primer lugar, se puede apreciar que para una sola etapa, la probabilidad de error es igual en ambos sistemas puesto a que el esquema de comunicación es idéntico y no existe diferencia entre ambos sistemas en la manera de procesar el símbolo recibido. \\

Al ir aumentando el número de etapas, las curvas de probabilidad de error de ambos sistemas decrecen de manera más lenta. Esto es consistente debido a que una mayor cantidad de etapas implica que la probabilidad de cometer un error en la decisión es mayor. \\

Para cualquier n, la probabilidad de error siempre es mayor en el digital hasta que las curvas de cada sistema se intersecan. De ahí en adelante, es mayor la del sistema analógico. Los puntos de intersección son los siguientes: \\

>n | SNR [db] | $P_e$
>--- | --- | ---
>5 | -0,4441 |0,4371
>9 | 0,2849 | 0,4792
>13 | 0,5517 | 0,4931
>17 | 0,7820 | 0,4976
>21 | 1,0071 | 0,4991
>25 | 1,0245 | 0,4997

Luego, al aumentar la SNR, la probabilidad de error es menor. Esto se debe a que como la SNR es una razón entre energía de la señal y su ruido, a mayor SNR, significa que la energía de la señal de interés tiene una mayor influencia en la regla de decisión que la energía del ruido, es por eso que la probabilidad de error disminuye. \\
Para valores de SNR negativos, es decir, para los casos en que el ruido es mayor a la señal, la probabilidad de error es mayor en el digital que en el analógico, sin importar la cantidad de las etapas. Si bien las curvas difieren, no se puede apreciar una diferencia considerable para los valores de probabilidad puesto a que no superan al 1% en el peor de los casos. \\
La cantidad de etapas influye más en el sistema analógico, que en el digital. Para el sistema digital, al tener una regla de decisión en cada etapa el efecto de la atenuación de la señal y el ruido no se van acumulando de una etapa a la subsiguiente. \\
Para el caso analógico, el ruido de la etapa anterior se ve amplificado y afecta a las siguientes hasta llegar al receptor que es el que toma la decisión. La señal se degrada y el ruido se incrementa en cada etapa de amplificación, es por eso que la probabilidad de error es mayor en el analógico. Los sistemas analógicos son más sensibles incluso a pequeñas cantidades de distorsión o ruido, y por lo tanto se puede afirmar que un sistema de comunicación digital es más robusto.\\
Por lo visto anteriormente, para valores de SNR negativos como es el caso en donde se tienen canales de comunicación con mucha adición de ruido, ya sea por deterioro en la red de cableado, interferencias con otras señales, factores ambientales, entre otros, es conveniente el uso de un sistema analógico. Sin embargo, como se analizó con anterioridad, no presenta una diferencia sustancial con respecto al analógico. \\
Luego, para un numero de etapas mayor que 1 y para el resto de valores de SNR, siempre es conveniente el sistema digital sobre el analógico.
