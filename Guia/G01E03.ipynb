```python
import math as m
import numpy as np
import scipy as sp
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
#Simulo X3 y X4

#creo ensayos de X1 y X2
N = 20000

X1 = np.random.uniform(-2,2,N)
X2 = np.random.uniform(-2,2,N)

X3 = X1+X2
X4 = X1 + 2*X2
# Grafico X1 y X2
f = plt.figure()
plt.scatter(X1,X2,s=5000/N,c='red')
```

```python
# Grafico histograma y acumulada X3
g, (ax1,ax2) = plt.subplots(2,1)
n, bins, patches = ax1.hist(X3,100,density=True,histtype='stepfilled', alpha=0.4, label = "Histograma de X3")
n, bins, patches = ax2.hist(X3,100,density=True,histtype='stepfilled', alpha=0.4,cumulative=True,label="Acumulada de X3")
ax1.legend(loc='best')
ax2.legend(loc='upper left')
```

```python
# Grafico histograma y acumulada X4
g, (ax1,ax2) = plt.subplots(2,1)
n, bins, patches = ax1.hist(X4,100,density=True,histtype='stepfilled', alpha=0.4, label = "Histograma de X4")
n, bins, patches = ax2.hist(X4,100,density=True,histtype='stepfilled', alpha=0.4,cumulative=True,label="Acumulada de X4")
ax1.legend(loc='best')
ax2.legend(loc='upper left')
```
