## Ejercicio 2 - Histograma
Sea $X$ una variable aleatoria exponencial de parametro $\lambda$ ( $X \sim Exp(\lambda)$ )
1. Genere $N$ = $10^4$ muestras de $X$
2. Calcule la media y la varianza muestrales de $X$ y comparelas con las teoricas.
3. Construya el histograma de las muestras de $X$. Normalice el histograma para que tenga area 1 y compare la funcion obtenida con la funcion  de densidad de probabilidad teorica.

```python
import math as m
import numpy as np
import scipy as sp
from scipy import stats as sps
import matplotlib.pyplot as plt
```

Para generar una exponencial, busco una transformacion g tal que pueda generar una expoencial con una uniforme

$F_X(x)=  1- e^{(-\lambda x)}= u$

$X = g(u) = \frac{-\ln 1- u}{\lambda}$

```python
#VA exponencial
def vaexp(l,N):
    u = np.random.rand(int(N)) # genero vector u, uniforme 0,1, con cant realizaciones
    x = -1*np.log(1-u)/l
    return x
```

```python
# Parametros
N = 1e5
l=0.5
X = vaexp(l, N)
```

```python
# Calculo la media y la varianza muestral de X
ux = np.mean(X)
vx = np.var(X)

print("La media es",f' Ux =%.2f'%ux,'\n')
print("La varianza es",r'Vx=%.2f'%vx)
```

La media teorica es $\mathbb{E}[X]=\frac{1}{\lambda}=2$ y la varianza teorica es $\mathbb{V}[X]=\frac{1}{\lambda^2}$= 4

```python
#Grafico historgrama 
plt.style.use('seaborn-pastel')
f= plt.figure()

n, bins, patches = plt.hist(X,100,density = True) # 1000 es los bins
#density es que cuente los que caen en los bins y divide por el total, "Normalice el histograma para que tenga area 1"
#lo que devuelve en bins es los bordes de los bins creados y n es la altura

#dibujo la densidad teorica
plt.plot(bins,l*np.exp(-l*bins),linewidth = 2)

#Armo leyenda con esperanza y varianza
textstr = '\n'.join((r'$\mu=%.2f$'%ux , r'$\sigma^2=%.2f$'%vx))
f.text(0.7, 0.85, textstr, fontsize=12, verticalalignment='top')

#Titulo del plot
plt.title('Histograma de VA exponencial de parametro l=  %.2f'%l)
#plt.show()
```
