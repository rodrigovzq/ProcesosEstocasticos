```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sps
```

```python
# defino funcion densidad conjunta
ux = 1
uy = -2
vx = 1
vy = 1
p = .9
Ca = lambda a,x,y: ((np.exp( -1*((((x-ux)/vx)**2)+(((y-uy)/vy)**2)-((2*p*(x-ux)*(y-uy))/(vx*vy)))/(2*(1-(p)**2)) )/(2*np.pi*vx*vy*np.sqrt(1-(p)**2)))-a)
```

```python
stroke = 25e-4
ancho = 10
x = np.arange(-3,5,stroke)
y = np.arange(-5,1,stroke)

r, q = np.meshgrid(x,y)

z=Ca(0,r,q)

#grafico

plt.contour(r,q,z)
```

```python
# tengo que generar 1e4 vectores normales
x1 = np.random.normal(ux,np.sqrt(vx),int(5e2))
x2 = np.random.normal(uy,np.sqrt(vy),int(5e2))
X = [x1, x2]

r, q = np.meshgrid(x1,x2)

z=Ca(0,r,q)

#grafico

plt.contour(r,q,z)
```
